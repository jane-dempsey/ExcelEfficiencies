```python
not_external = ['MVOAY Expired', 'MVOAY Missing', 'NSOPW Missing', 'NSOPW Expired', 'CPS Exempt', 'MVOAY Expired;', 'MVOAY Missing;', 'NSOPW Expired;', 'NSOPW Missing;', 'CPS Exempt;']
```

```python
df['faultsupdate'] = df['faults'].str.replace(pat, '').str.replace('I9', 'IZ')
```

```python
df = df[['serviceid', 'branch', 'servicename', 'ContractStatus','cityspanid', 'faultsupdate', 'roles', 'staffname', 'programspecialists', 'programmanagers']]
df
```

```python
df = df.fillna('nothing')
```

```python
cannot_work = ['ASM Pending', 'CPS Pending,', 'DCFS Pending', 'ASM Missing', 'CPS Missing', 'IZ Missing', 'DCFS Missing', 'MR Cert Missing']
missing_items = ['Confidential Missing', 'Contact Info Missing', 'Education Form Missing', 'Emp Handbook Missing', 'Experience Form Missing', 'MR Ack Missing', 'MR Ack Missing', 'MR Ack Expired', 'PACR (IND) Missing', 'Signed Budget Missing', 'Social Media Policy Missing']
```

```python
def match_list(faultsupdate, staffname):
    k = ([cannot_work[cannot_work.index(s)] for s in cannot_work if s in faultsupdate]) or ([s for s in cannot_work if s in faultsupdate])
    return k[0] if k else ''

df['notworkstatus'] = df.apply(lambda x: match_list(x['faultsupdate'], x['staffname']), axis=1)
```

```python
cw_cleanup = r'\b(?:{})\b'.format('|'.join(cannot_work))
```

```python
df['notworkstatus'] = df['notworkstatus'].str.replace(cw_cleanup, 'Cannot Start - See Details for Missing Items')
```

```python
df[:4]
```

```python
def match_list_canworkmiss(faultsupdate, staffname):
    k = ([missing_items[missing_items.index(s)] for s in missing_items if s in faultsupdate]) or ([s for s in missing_items if s in faultsupdate])
    return k[0] if k else ''

df['workstatuscanworkmiss'] = df.apply(lambda x: match_list_canworkmiss(x['faultsupdate'], x['staffname']), axis=1)
```

```python
miss_cleanup = r'\b(?:{})\b'.format('|'.join(missing_items))
```

```python
df['workstatuscanworkmiss'] = df['workstatuscanworkmiss'].str.replace(miss_cleanup, 'Can Start with Missing Items - See Details for Missing Items')
```

```python
df[:4]
```

```python
no_working = ['Cannot Start - See Details for Missing Items']
yes_working_but = ['Can Start with Missing Items - See Details for Missing Items']
```

```python
df['Can Staff Start Work?'] = np.where(df['notworkstatus'].str.contains('|'.join(no_working)),
        'NO CANNOT WORK - see details',
         np.where(df['workstatuscanworkmiss'].str.contains('|'.join(yes_working_but)),
             'Yes but missing items -- see details',
             'Yes can work, nothing else needed this session'))
```
